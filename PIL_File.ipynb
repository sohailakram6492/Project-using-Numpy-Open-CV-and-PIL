```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.title("1st")
plt.imshow(coppied_image)
plt.subplot(2,2,2)
plt.title("2nd")
plt.imshow(coppied_image)
plt.subplot(2,2,3)
plt.title("3rd")
plt.imshow(coppied_image)
plt.subplot(2,2,4)
plt.title("4th")
plt.imshow(coppied_image)
```

```python
from PIL import ImageFont
from PIL import ImageDraw
watermarked_image=image.copy()
draw=ImageDraw.Draw(watermarked_image)
font=ImageFont.truetype("arial.ttf",100)
draw.text((0,0),"xyz",(0,0,0),font=font)
plt.imshow(watermarked_image)
```

```python
size=(200,100)
crop_image=image.copy()
crop_image.thumbnail(size)
coppied_image=image.copy()
coppied_image.paste(crop_image,(600,200))
plt.imshow(coppied_image)
```

```python
bw_image=image.convert("L")
plt.imshow(bw_image,cmap="gray")
```

```python
new_formated_image=image.convert('HSV')
new_formated_image.mode
```

```python
numpy_array=np.array(image)
numpy_array
```

```python
numpy_array=Image.fromarray(numpy_array)
plt.imshow(numpy_array)
```

```python
from PIL import ImageEnhance
image_copy=image.copy()
image_enhance=ImageEnhance.Color(image_copy).enhance(5.5)
plt.imshow(image_enhance)
```

```python
image_channel=image.copy()
r,g,b=image_channel.split()
im=Image.merge("RGB",(b,g,r))
plt.imshow(im)
```

```python
size=(220,320)
crop_image=image.copy()
crop_image=crop_image.resize(size)
coppied_image=image.copy()
coppied_image.paste(crop_image,(610,120))
plt.imshow(coppied_image)
```
